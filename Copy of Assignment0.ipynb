```python
#Problem 1 

import numpy as np 
import matplotlib.pyplot as plt 

file = open('test.txt', 'r')
char1 = file.read() 
char2 = sorted(char1.lower())    
# print(char1)

def char_frequency(char2):
    dict = {}
    for n in char2:
      if n.isalpha():
        keys = dict.keys()
        if n in keys:
            dict[n] += 1
        else:
            dict[n] = 1
    return dict

counts = char_frequency(char2)
counts2 = sorted(counts)
print(counts)

alphabets = np.array(list(counts.keys()))
frequency = np.array(list(counts.values()))

print(frequency)

norm = [float(i)/sum(frequency) for i in frequency]

plt.bar(alphabets,norm)
plt.title("Normalized bar graph")
plt.xlabel("Alphabets")
plt.ylabel("Frequency")
plt.show()

plt.bar(alphabets,frequency)
plt.title("Unnormalized bar graph")
plt.xlabel("Alphabets")
plt.ylabel("Frequency")
plt.show()
```

```python
#Problem 2

import numpy as np
import cv2
from PIL import Image
img = cv2.imread('test.png')
img3 = cv2.imread('test.png')
img4 = cv2.imread('test.png')
# print(img)
#print(img[0,0,:])
def leastdist(a):
    x = np.array((255, 0, 0))
    y = np.array((0, 255, 0))
    z = np.array((0, 0, 255))
    p = np.linalg.norm(x-a)
    q = np.linalg.norm(y-a)
    r = np.linalg.norm(z-a)
    if min(p,q,r) == p:
       return x
    if min(p,q,r) == q:
       return y
    if min(p,q,r) == r:
       return z

for i in range(0,127):
  for j in range (0,127):
     img3[i][j] = leastdist(img[i][j])

for i in range (39,88):
  for j in range (39,88):
     img4[i][j] = [0,0,0]


im = Image.fromarray(img3)
im.save("output1.png")

im = Image.fromarray(img4)
im.save("output2.png")
```
